```python
#import important libraries
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
```

```python
ds.shape
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_rows',None)

ds.head(6)
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
miss_value_percent=ds.isnull().sum()/ds.shape[0]*100   #%of missing values
miss_value_percent
```

```python
drop_columns=miss_value_percent[miss_value_percent>17].keys()
drop_columns
```

```python
ds2=ds.drop(columns=drop_columns) #new dataframe df2 after column drop
```

```python
ds2.shape
```

```python
ds2.select_dtypes(include=['int64','float64']).columns#only numerical columns
```

```python
ds3=ds2.select_dtypes(include=['int64','float64'])
ds3.head(6)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(ds3)
```

```python
ds3[ds3.isnull().any(axis=1)]
```

```python
ds3.isnull().sum()
```

```python
miss_num_var=[var for var in ds3.columns if ds3[var].isnull().sum()>0]
miss_num_var
```

```python
plt.figure(figsize=(16,9))

sns.set()
for i ,var in enumerate(miss_num_var):
    plt.subplot(1,2,i+1)
    sns.distplot(ds3[var],bins=20,kde_kws={'linewidth':5,'color':'k'})
```

```python
ds4=ds3.fillna(ds3.mean())   #NaN value filled by mean value
ds4.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))

sns.set()
for i ,var in enumerate(miss_num_var):
    plt.subplot(1,2,i+1)
    sns.distplot(ds3[var],bins=20,kde_kws={'linewidth':8,'color':'k'},label='original')
    sns.distplot(ds4[var],bins=20,kde_kws={'linewidth':5,'color':'y'},label='mean')
    
    plt.legend()
```

```python
ds5=ds3.fillna(ds3.mean())   #NaN value filled by median value
ds5.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))

sns.set()
for i ,var in enumerate(miss_num_var):
    plt.subplot(1,2,i+1)
    sns.distplot(ds3[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'k'},label='original')
    sns.distplot(ds4[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'c'},label='mean')
    sns.distplot(ds5[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'r'},label='median')
    plt.legend()
```
